# Default Normal Deep Neural Network:
- Kernel_init --> he normal
- activation function --> ReLu, elu
- Normalization --> None, If shallow NN but if deep NN use BN
- Regularization --> early stoping, L2 only if needed
- Optimizers --> momentum, Adam, RMSprop



# Complex Deep Neural Network:
- Regularization --> dropout
- optimizer --> Adam
- activation function --> ReLu, selu
- initialization --> Lecun initialization
